# 필터링한 엑셀을 사용해서 여기서부터 해주세요!!!

```python
# 도시고속도로, 월~금, 차선 수 3개 이상
df1 = pd.read_csv('2020년 8월.csv', encoding = 'UTF-8')
df1
```

```python
# 1차 필터링 이후 도로 수
len(df1['링크아이디'].unique())
```

```python
# 칼럼이름들 확인하기
df1.columns
```

```python
# 요일과 링크아이디의 조합 만들기 > 비정형데이터 부분 살리기
df2 = df1[['요일', '도로명', '링크아이디', '시점명', '종점명', '방향', '거리',
           '차선수', '도심/외곽구분', '권역구분']]
df2
```

```python
# 링크아이디별와 요일별 시간별 평균 값 구하기
df1.drop(['일자','기능유형구분'], axis = 1, inplace = True)
d#f1 = df1.groupby(['요일', '링크아이디']).mean()
df1
```

```python
# 요일과 링크아이디의 조합
df2 = df2.drop_duplicates().reset_index(drop = True)
df2
```

```python
# 결과 df 나오는 부분
df_result = pd.merge(df2, df1, on = ['요일', '링크아이디'])
df_result
```

```python
df_result.columns
```

```python
# 컬럼명 정리하기
df_result.drop(['거리_y', '차선수_y'], axis = 1, inplace = True)
df_result.rename(columns = {'거리_x' : '거리', '차선수_x' : '차선수'}, inplace = True)
```

```python
# 결과파일 csv로 저장하기
df_result.to_csv("data_08.csv", encoding = 'cp949')
```
