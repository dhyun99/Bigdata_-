```python
import pandas as pd
import numpy as np
import matplotlib as plt
import openpyxl
```

```python
# sheet name 추출
import openpyxl

wb = openpyxl.load_workbook('월간교통량_속도_평균_상하행.xlsx')
ws_names = wb.sheetnames

sheet_name = []
for i in ws_names:
    sheet_name.append(i)
    
sheet_name
```

```python
wb = openpyxl.Workbook()
new_file = '연간교통량_속도_평균_상하행_비율_하상.xlsx' #새로 만들 파일 이름
wb.save(new_file)
```

```python
writer = pd.ExcelWriter(new_file, engine = 'xlsxwriter')
```

```python
#월 평균 속도 데이터를 비율로 만들기, 시트별 저장
for i in sheet_name:
    df = pd.read_excel('월간교통량_속도_평균_상하행.xlsx', sheet_name = i)   
    
    df2 = df.sort_values(by=['링크아이디'], axis=0) #링크아이디 순 정렬
    
    tmp= df2[['요일','도로명', '링크아이디', '시점명', '종점명', '방향', '거리', '차선수', '도심/외곽구분', '권역구분']]
    
    tmp2=df2.groupby(['링크아이디']).mean()
    
    df3 = pd.merge(tmp,tmp2, on = ['링크아이디'])
    
    df_final = df3.loc[df3['요일']=='월',]   # 월요일만 
    
    df_final.drop(['요일','Unnamed: 0','거리_y','차선수_y'], axis = 1, inplace = True)
    
    df_final_up = df_final.loc[df_final['방향']=='상행',]   # 상행만
    df_final_down = df_final.loc[df_final['방향']=='하행',]  # 하행만
    
    df_ratio_up = pd.DataFrame.copy(df_final_up) # 속도비율 데이터용 상행데이터 복사 
    df_ratio_down = pd.DataFrame.copy(df_final_down) # 속도비율 데이터용 하행데이터 복사 
    
    #속도비율 데이터 만들기
    for j in range(df_final_up.shape[0]):
        for k in range(24):
            up,down = df_final_up.iloc[j,k+9],df_final_down.iloc[j,k+9]
            df_ratio_up.iloc[j,k+9] = up/down #상행/하행->1이상이면 상행이 빠름, 1이하면 상행이 느림
            df_ratio_down.iloc[j,k+9] = down/up #하행/상행->1이상이면 하행이 빠름, 1이하면 하행이 느림
    
    
    # 2. 시트별 데이터 추가하기(1개 시트로 저장할때와 유사) 
    df_ratio_down.to_excel(writer, sheet_name= i)
    #df_ratio_down.to_excel(writer, sheet_name= i)
    
# 3. 엑셀 파일 저장하기
writer.save()
```

```python
df_real_final= pd.concat([df0,df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11])
```

```python
#링크아이디 중복 문제를 12개씩 인덱스로 뽑음-->중복 제거
df_real_final2= df_real_final[['도로명', '링크아이디', '시점명', '종점명', '방향', '거리_x', '차선수_x', '도심/외곽구분', '권역구분']]
    
df_real_final3=df_real_final.groupby(['링크아이디']).mean()
    
df_real_real = pd.merge(df_real_final2,df_real_final3, on = ['링크아이디'])

index_list=[]
for i in range(0,816,12):
    index_list.append(i)
    
pd.set_option('display.max_columns', None)
df_real_real2=df_real_real.loc[index_list]

df_real_real2.drop(['Unnamed: 0','거리_x_y','차선수_x_y'], axis = 1, inplace = True)
#df_final.drop(['요일','Unnamed: 0','거리_y','차선수_y'], axis = 1, inplace = True)
```
