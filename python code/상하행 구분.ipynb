```python
# sheet name 추출
import openpyxl

wb = openpyxl.load_workbook('월간교통량_평균.xlsx')
ws_names = wb.sheetnames

sheet_name = []
for i in ws_names:
    sheet_name.append(i)
    
sheet_name
```

```python
wb = openpyxl.Workbook()
new_file = '월간교통량_평균_상행하행.xlsx'
wb.save(new_file)
```

```python
writer = pd.ExcelWriter(new_file, engine = 'xlsxwriter')
```

```python
for i in sheet_name:
    df = pd.read_excel('월간교통량_평균.xlsx', sheet_name = i)   
    
    dd = df.sort_values(by=['링크아이디'], axis=0)
    
    mon = dd[dd['요일'] == '월']
    tue = dd[dd['요일'] == '화'] 
    wed = dd[dd['요일'] == '수'] 
    thu = dd[dd['요일'] == '목'] 
    fri = dd[dd['요일'] == '금'] 
    
    df_final=pd.concat([mon,tue,wed,thu,fri])
    
    # 2. 시트별 데이터 추가하기(1개 시트로 저장할때와 유사) 
    df_final.to_excel(writer, sheet_name= i)
    
# 3. 엑셀 파일 저장하기
writer.save()
```
